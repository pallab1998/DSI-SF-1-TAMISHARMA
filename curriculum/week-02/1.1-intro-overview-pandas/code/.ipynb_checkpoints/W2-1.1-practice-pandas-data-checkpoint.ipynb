#### 1. Load the Mad Men cast show data into a pandas DataFrame

#### 4. Rename any columns with spaces or special characters to not have them

```python
special_characters = ['$','#','?','/',' ']
```

```python
for col in cast_d.columns:
    if any([x in col for x in special_characters]):
        newcol = ''.join([x for x in col if x not in special_characters])
        cast_d.rename(columns={col:newcol}, inplace=True)
```

```python
cast_d.head()
```

#### 5. Subset the data where the status of the show is not "END" or "End"

```python
mask = (cast_d['Status'] != 'END') & (cast_d['Status'] != 'End')
subset = cast_d[mask]
```

```python
subset.head()
```

#### 6. Print out the performers where the show start is greater than 2005 and the score is greater than 7.

```python
score_mask = (cast_d['ShowStart'] > 2005) & (cast_d['Score'] > 7)
subset = cast_d[score_mask]
subset.head()
```
