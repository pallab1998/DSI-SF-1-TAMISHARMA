## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. 

```python
billboard.head()
```

```python
#removing unnecessary x in front of weeks and removing dots

billboard.columns = billboard.columns.map(lambda x: x.replace('x', ''))
billboard.columns = billboard.columns.map(lambda x: x.replace('.', ''))
billboard.head()
```

```python
billboard.rename(columns = {'artistinverted': 'artist_inverted', 'dateentered': 'date_entered',
                            'datepeaked': 'date_peaked', '13thwek': '13thweek', 'time':'length'},
                inplace = True)
billboard['1stweek'] = billboard['1stweek'].apply(float)
billboard.head()
```

```python
###According to:  http://www.billboard.com/charts/hot-100/2000-01-01 and http://www.billboard.com/artist/431329/whitney-houston/chart?page=1&f=379
###My Love Is Your Love peaked on January 1 2000

billboard[billboard["track"] == "My Love Is Your Love"]
billboard.iloc[27,6] = '2000-01-01'
billboard.info()
```

```python
# accessing specific month in billboard
# for value in billboard['date_entered']:
#     if value.month == 2:
#         print 'february'
```

```python
billboard['date_entered'] = pd.to_datetime(billboard['date_entered'])
billboard['date_peaked'] = pd.to_datetime(billboard['date_peaked'])
billboard.dtypes.head(10)
```

```python
# New column of how many weeks a song is on the charts, 
# Another new column of how many days a song takes to peak,
# And dropping the columns with no numerical values.

billboard['weeks_on_chart_count'] = billboard.iloc[:,7:72].count(
    axis=1, level=None, numeric_only=True)

billboard["days_to_peak"] = (billboard["date_peaked"] - billboard["date_entered"]).dt.days

crap_data = ['66thweek', '67thweek', '68thweek', '69thweek', '70thweek',
             '71stweek', '72ndweek', '73rdweek', '74thweek', '75thweek', '76thweek']
billboard.drop(crap_data, axis = 1, inplace = True)

billboard.head()
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
weeks = list(billboard.ix[:,7:72
                         ])
bill_long = pd.melt(billboard,
                        id_vars = ['artist_inverted', 'track','length', 'genre',
                                   'date_entered', 'date_peaked', 'days_to_peak', 'weeks_on_chart_count'],
                        value_vars = weeks,
                        var_name = 'week',
                        value_name = 'ranking')

bill_long.head()
```

```python
### Convert 'week' values to an int for aggregate function applications.
bill_long['week'] = bill_long['week'].map(lambda x: x.replace('thweek', ''))
bill_long['week'] = bill_long['week'].map(lambda x: x.replace('stweek', ''))
bill_long['week'] = bill_long['week'].map(lambda x: x.replace('ndweek', ''))
bill_long['week'] = bill_long['week'].map(lambda x: x.replace('rdweek', ''))
bill_long['week'].unique()
```

```python
bill_long['week'] = bill_long['week'].apply(int)
bill_long.dtypes
```

```python
bill_long.shape
```

```python
grouped_long = bill_long.groupby(['artist_inverted', 'track', 'week', 'ranking'])
grouped_long.agg({'week': max}).head(25)
```

```python
billboard["artist_inverted"].value_counts().head(10)
```

```python
billboard["genre"].value_counts().head()
```

```python
# accessing specific month in billboard
# for value in billboard['date_entered']:
#     if value.month == 2:
#         print 'february'

### adding in a column for the month that the song peaks in

billboard["peak_months"] = [value.month for value in billboard["date_peaked"]]
bill_long["peak_months"] = [value.month for value in bill_long["date_peaked"]]
```
